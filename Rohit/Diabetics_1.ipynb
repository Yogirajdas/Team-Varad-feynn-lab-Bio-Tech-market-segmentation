```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
X1 = df.loc[:, ["Age", "Diabetic"]].values
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
WCSS = []

# Use k from 1 to 15
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k, init='random', n_init=10, max_iter=500,  tol=1e-04, random_state=0)
    km = km.fit(X1)
    # Get sum of square distances by applying km.inertia_ 
    Sum_of_squared_distances.append(km.inertia_)

# Plot Results
plt.plot(K, Sum_of_squared_distances, marker='o')
plt.xlabel('k')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)

label1 = kmeans.fit_predict(X1)

print(label1)
```

```python
fig = plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x="Age", y="Diabetic", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],marker="X", c="r", s=80, label="centroids") 
plt.legend()
plt.show()
```
